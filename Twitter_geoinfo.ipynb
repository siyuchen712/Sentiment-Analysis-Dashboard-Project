```python
def excludeword(words):
    wordsFiltered = []
    for w in words:
        if w not in stopWords_all:
            wordsFiltered.append(w)
    return wordsFiltered
```

```python
words = []
wordsFiltered = []

for i in range(len(data_update)):
    words = words + tokenizer.tokenize(re.sub(r"http\S+", '', data_update[i]['full_text']))
    words = [word.lower() for word in words]
    for w in words:
        if w not in stopWords_all:
            wordsFiltered.append(w)
```

```python
wordfreq = json.dumps(nltk.FreqDist(wordsFiltered))
dict_wordfreq = json.loads(wordfreq.replace("'", "\""))
word_ls = list(dict_wordfreq.keys())
```

```python
sent_rate_ls = []
final_word_ls = []

for word in word_ls:
    for tweet in data:
        sentence_word = excludeword(tokenizer.tokenize(re.sub(r"http\S+", '',tweet['full_text'].lower())))
        if word in sentence_word:
            sent_rate_ls.append(tweet['sentiment_rate'])
        else: continue

    word_sentiment = sum(sent_rate_ls) / float(len(sent_rate_ls))
    final_word_ls.append([word, dict_wordfreq[word], word_sentiment])
```

```python
word_word = []
word_count = []
word_sentiment_rate = []
    
for i in final_word_ls:
    word_word.append(i[0])
    word_count.append(i[1])
    word_sentiment_rate.append(i[2])
    
df_wordfreq = pd.DataFrame({'word': word_word, 'count': word_count, 'sentment rate': word_sentiment_rate})

df_wordfreq.to_csv('wordfreq_wpi_cleaned.csv', sep=',')
```

```python
def getplace(lat, lon):
    attempts = 0
    success = False
    url = "http://maps.googleapis.com/maps/api/geocode/json?"
    url += "latlng=%s,%s&sensor=false" % (lat, lon)
    while success != True and attempts < 3:
        response = urlopen(url).read()
        attempts += 1
        j = json.loads(response)
        if (j['status'] == "OVER_QUERY_LIMIT"):
            time.sleep(2)
            continue
        success = True
        
    if (j['status'] == "OK"):
        components = j['results'][0]['address_components']
        country = city = state = None
        for c in components:
            if "country" in c['types']:
                country = c['long_name']
            if "administrative_area_level_1" in c['types']:
                state = c['long_name']    
            if "locality" in c['types']:
                city = c['long_name']

    else:
        city = ''
        state = ''
        country = ''

    return city, state, country
```

```python
loc_google_map = []
j = 0
for i in data_update:
    if i['latitude'] == '' or i['longitude'] == '' or i['latitude'] is None or i['longitude'] is None:
        loc_google_map.append([])
    else:
        haha = list(getplace(i['latitude'], i['longitude']))
        loc_google_map.append(haha)
    print(j)
    j = j+1
```

```python
j = 0
for i in data_update:
    if len(loc_google_map[j]) == 0:
        i['city'] = ''
        i['state'] = ''
        i['country'] = ''
    else:
        i['city'] = loc_google_map[j][0]
        i['state'] = loc_google_map[j][1]
        i['country'] = loc_google_map[j][2]
    j = j+1
```
